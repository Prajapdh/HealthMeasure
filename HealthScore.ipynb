```python
#https://www.geeksforgeeks.org/working-zip-files-python/

import os
import shutil
from zipfile import ZipFile

DIABETES_PATH = "D:\Machine Learning projects\Make UC\datasets/diabetes"
FILE_NAME = "diabetes-health-indicators-dataset.zip"
def extract_diabetes_data(file_name=FILE_NAME,diabetes_path=DIABETES_PATH):
    if not os.path.isdir(diabetes_path):
        os.makedirs(diabetes_path)
#     shutil.move(file_name, diabetes_path)
    with ZipFile(file_name,'r') as zip:
        zip.printdir()
        zip.extractall(diabetes_path)

extract_diabetes_data()
```

```python
def load_diabetes_data(diabetes_path=DIABETES_PATH):
     csv_path = os.path.join(diabetes_path, "diabetes_012_health_indicators_BRFSS2015.csv")
     return pd.read_csv(csv_path)
diabetes = load_diabetes_data()
diabetes.info()
```

```python
del diabetes['Education']
del diabetes['Income']
del diabetes['NoDocbcCost']
del diabetes['AnyHealthcare']
del diabetes['CholCheck']
del diabetes['Age']
```

```python
filtered_indices = np.where((diabetes['Diabetes_012']!=2))
diabetes = diabetes.loc[filtered_indices]
```

```python
def split_train_test(data_frame,test_ratio):
    shuffled_indices = np.random.permutation(len(data_frame))
    test_set_len = int(len(data_frame)*test_ratio)
    test_set_indices = shuffled_indices[:test_set_len]
    train_set_indices = shuffled_indices[test_set_len:]
    return data_frame.iloc[train_set_indices], data_frame.iloc[test_set_indices]

train_set, test_set = split_train_test(diabetes,0.2)
# print("train set len: ",len(train_set),", test set len: ",len(test_set))
```

```python
def export_test_data(data_frame,diabetes_path=DIABETES_PATH):
     csv_path = os.path.join(diabetes_path, "diabetes_test_set.csv")
     return data_frame.to_csv(csv_path)

export_test_data(test_set)

# demo = pd.read_csv("D:\Machine Learning projects\Make UC\datasets/diabetes\diabetes_test_set.csv")
# demo.head()
```
